# Simple Linear Regression. Minimal example

### Import the relevant libraries

```python
# We must always import the relevant libraries for our problem at hand. NumPy is a must for this example.
import numpy as np

# matplotlib and mpl_toolkits are not necessary. We employ them for the sole purpose of visualizing the results.  
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
observations = 1000

xs=np.random.uniform(-10,10, size=(observations,1))
zs=np.random.uniform(-10,10, (observations,1))
```

```python
inputs = np.column_stack((xs,zs))
```

```python
noise= np.random.uniform(-1,1,(observations,1))
```

```python
targets = 2*xs -3*zs + 5 + noise
targets.shape
```

```python
targets = targets.reshape(observations,)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(xs, zs, targets)
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('targets')
ax.view_init(azim=100)
plt.show()
```

```python
targets = targets.reshape(observations,1)
```

```python
init_range =0.1
weights = np.random.uniform(-init_range, init_range, size=(2,1))
bias =np.random.uniform(-init_range, init_range, size =1)
```

```python
learning_rate =0.01
```

```python
for i in range(1000):

    outputs = inputs@weights + bias
    deltas = outputs -targets
    loss =np.sum(deltas ** 2)/2/observations
    print(loss)

    deltas_scaled =deltas/observations
    weights = weights - learning_rate*inputs.T@deltas_scaled
    bias = bias - learning_rate*np.sum(deltas_scaled)
```

```python
print(weights, bias)
```

```python
plt.figure(dpi=200)
plt.scatter(targets, outputs, alpha =0.3)
plt.plot(targets, targets, color ='red', linestyle='-.')
```
