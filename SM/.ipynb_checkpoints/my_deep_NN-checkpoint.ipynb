```python
npz  = np.load('Audiobooks_train_data.npz')
```

```python
train_inputs = npz['inputs'].astype(np.float32)
```

```python
train_targets = npz['targets'].astype(np.int32)
```

```python
npz_val  = np.load('Audiobooks_val_data.npz')
val_inputs = npz_val['inputs'].astype(np.float32)
val_targets = npz_val['targets'].astype(np.int32)
```

```python
npz_test  = np.load('Audiobooks_test_data.npz')
test_inputs = npz_test['inputs'].astype(np.float32)
test_targets = npz_test['targets'].astype(np.int32)
```

```python
input_size = 10
output_size =2
hidden_layer_size =50

model =tf.keras.Sequential([
                            tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
                            tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
                            tf.keras.layers.Dense(output_size, activation='softmax'),
    
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy']  )
batch_size =200
max_epochs =100

early_stopping = tf.keras.callbacks.EarlyStopping(patience=2)
```

```python
model.fit(train_inputs, 
          train_targets, 
          batch_size = batch_size, 
          epochs= max_epochs, 
          validation_data=(val_inputs, val_targets), 
          callbacks = [early_stopping],
          verbose =2 )
```

```python
test_loss, test_accuracy = model.evaluate(test_inputs, test_targets)
```
